## Get The Data ##

```python
ROOT = 'http://www.csun.edu/~hceng029/yeats/'
data = urllib2.urlopen(ROOT+'collectedpoems.html')
```

```python
doc = BeautifulSoup(data)
```

```python
book_links = [ROOT +link.get('href') for link in doc.find_all('a') if 'poems' in link.get('href')]
book_titles = [l.text for l in doc.find_all('a')]
```

```python
def build_corpus(book_links,book_titles):    
    corpus = defaultdict(dict) 
    
    for book_link, book_title in itertools.izip(book_links,book_titles):
        u = urllib2.urlopen(book_link)
        soup = BeautifulSoup(u)
        poem_root = 'http://www.csun.edu/~hceng029/yeats/yeatspoems/'
        poem_links = [poem_root+link.get('href') for link in soup.find_all('a') ]
        poem_titles = [p.text for p in soup.find_all('a')]
        
        for poem_link, poem_title in itertools.izip(poem_links,poem_titles): 
            text = urllib2.urlopen(poem_link).read()
            text= ' '.join(text.split('\n')[2:])
            corpus[book_title][poem_title] = text
    return corpus
    
    
```

```python
corpus = build_corpus(book_links,book_titles)
```

The text then needs to be processed using NLTK in order to: 
- remove case
- remove stopwords
- remove punctuation 
- tokenize words (split words apart) 
- stem words i.e make 'bird' and 'birds' the same 

We will define a function that will bundle these preprocessing steps, which can then be passed to the various scikit vectorizers 



```python
from nltk import stem
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

porter = stem.porter.PorterStemmer()

def preprocess(text):
    sentence = text.lower()
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(sentence)
    filtered_words = [w for w in tokens if not w in stopwords.words('english')]
    stemmed = [porter.stem(word) for word in filtered_words]
    return " ".join(stemmed)
```
